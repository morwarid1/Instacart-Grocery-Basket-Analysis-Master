### Question 2 ) aggregatoion for the entire dataframe

```python
#Performing a single Aggregation - 
#spliting the data into groups based  on department_id 
# applying the agg() function to each group to obtain the mean values of "order_number" 
df.groupby('department_id').agg({'order_number':['mean']})
```

```python
# performing multiple aggregations
df.groupby('department_id').agg({'order_number': ['mean','min','max']})
```

```python
df.head(100)
```

```python
# trying the aggregation with the entire dataframe 
# performing multiple aggregations
df_orders_products_merged.groupby('department_id').agg({'order_number': ['mean','min','max']})
```

### Question 4 ) Create loyalty flag using transfomr and loc

```python
# Creating loyalty flag
# spliting data into groups based on the "user_id"
# apply the transform function on the "order_number" to find max order
# create new column "max_order" into which you place the result of aggregation
df_orders_products_merged['max_order'] = df_orders_products_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_orders_products_merged.head(100)
```

```python
# set the flag using loc
# max orders over 40 = loyal customer
df_orders_products_merged.loc[df_orders_products_merged['max_order']>40,'loyalty_flag']='loyal customer'
```

```python
# max order over 10 but less or equal 40 =regular customer
df_orders_products_merged.loc[(df_orders_products_merged['max_order']<=40)& (df_orders_products_merged['max_order']>10),'loyalty_flag']='Regular customer'
```

```python
# max order less than or equal to 10 = new customer 
df_orders_products_merged.loc[df_orders_products_merged['max_order']<= 10,'loyalty_flag'] = 'New customer'
```

```python
# printing the frequency of loyalty flag 
df_orders_products_merged['loyalty_flag'].value_counts(dropna = False)
```

### Question 6 ) finding the different types of spenders

```python
# looking at the price of item 
# createing a spending flag 
```

```python
# looking at the price column values
df_orders_products_merged.head(1000000)
```

```python
# aggregating Data
df_orders_products_merged['average_prices'] = df_orders_products_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# max order less than or equal to 10 = new customer 
df_orders_products_merged.loc[df_orders_products_merged['average_prices']<10,'spending_flag']='low spender'
```

```python
# max order less than or equal to 10 = new customer 
df_orders_products_merged.loc[df_orders_products_merged['average_prices']>=10,'spending_flag']='high spender'
```

```python
# printing the frequency of spending_flag
df_orders_products_merged['spending_flag'].value_counts(dropna = False)
```

### Question 7 ) Determine frequent versus non-frequent customers and regular customer

```python
# aggregating Data
df_orders_products_merged['ordering behavior'] = df_orders_products_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# max order less than or equal to 10 = new customer 
df_orders_products_merged.loc[df_orders_products_merged['ordering behavior']>20,'order_frequency_flag']='Non-frequent customer'
```

```python
# max order over 10 but less or equal 40 =regular customer
df_orders_products_merged.loc[(df_orders_products_merged['ordering behavior']>10)& (df_orders_products_merged['max_order']<=20),'order_frequency_flag']='Regular customer'
```

```python
# max order less than or equal to 10 = new customer 
df_orders_products_merged.loc[df_orders_products_merged['ordering behavior']<=10,'order_frequency_flag']='Frequent customer'
```

```python
# printing the frequency of spending_flag
df_orders_products_merged['order_frequency_flag'].value_counts(dropna = False)
```
