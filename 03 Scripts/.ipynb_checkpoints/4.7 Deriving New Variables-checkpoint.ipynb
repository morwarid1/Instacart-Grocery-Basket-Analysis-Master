### Question 1 ) Creating the " price_label" and "busiest_day"

```python
# creating subset of the dataframe 
df = df_orders_products_merged[:1000000]
```

```python
df.shape
```

```python
# defining price labeles
def price_label (row): 
    
    if row ['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices']>5) and (row['prices']<= 15):
        return 'Mid-range product'
    elif row ['prices']>15:
        return 'high range'
    else: return 'Not enough data'
```

```python
# applying the function price lable
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# create a subset
df ['price_range'].value_counts(dropna= False)
```

```python
# checking the most expensive product
df['prices'].max()
```

```python
# product bought on the busiest and slowest days- 
#finding frequency of "orders_day_of_the_week" it saved as order_dow
df_orders_products_merged['order_dow'].value_counts(dropna=False)
```

```python
#Createing new column as "busiest day","least busy" and "regularly busy"

result = []

for value in df_orders_products_merged["order_dow"]:
    if value == 0:
        result.append("Busiest days")
    elif value == 4:
        result.append("slowest days")
    else:
        result.append("Regularly busy")
```

```python
result
```

### Question 2 ) Creating labels for "Busiest days"and two slowest days

```python
# busiest days
df_orders_products_merged['busiest days'] = result
```

```python
# slowest days
df_orders_products_merged["slowest days"]=result
df_orders_products_merged["slowest days"].value_counts(dropna =False)
```

### Question 4 ) Identifying" Most orders" "Average orders" and "Fewest orders" . Create new column "busiest_period_of_day"

```python
# product bought on the busiest period of the day / busiest hour of the day

df_orders_products_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
#Createing new column as "Most orders","Averege orders" and "Fewest orders"

result = []

for value in df_orders_products_merged["order_hour_of_day"]:
    if value == 10:
        result.append("Most orders")
    elif value == 3:
        result.append("Fewest orders")
    else:
        result.append("Averege orders")
```

```python
result
```

```python
# Most orders
df_orders_products_merged['Most orders'] = result
```

```python
# Averege orders  n
df_orders_products_merged['Averege orders'] = result
```

```python
# Fewest orders
df_orders_products_merged['Fewest orders'] = result
```
